```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
log_pro = pd.read_csv('All_times_Pro_logged_3UTR_3mer.txt', sep='\t', index_col=0)
log_pro = log_pro[log_pro.columns[2:len(log_pro.columns)]]
```

```python
log_pro
```

```python
log_pro
arr=[]
for i in log_pro.loc['CAT']:
    arr.append(2**i)
```

```python
arr
```

```python
df= pd.DataFrame(index=log_pro.loc['CAT'].index)
df['Proportion of Weighted Occurance']= arr
```

```python
df
```

```python
df['Log of Proportions of Weighted Occurances']=log_pro.loc['CAT']
```

```python
df
```

```python
df['zscore']= st.zscore(log_pro.loc['CAT'])
```

```python
df
```

```python
df['zscore']= st.zscore(log_pro.loc['CAT'])
df.loc['Mean']= df[['Proportion of Weighted Occurance','Log of Proportions of Weighted Occurances']].mean()
df.loc['STD']= df[['Proportion of Weighted Occurance','Log of Proportions of Weighted Occurances']].std()
```

```python
df
```

```python
fig = plt.figure(figsize=(20, 20))
median = pd.Series(arr).median()
plt.hist(df.loc[log_pro.columns]['Proportion of Weighted Occurance'], bins=50, edgecolor='black', color = 'red')

plt.title("Frequency of the Proportions for CAT in the 3'UTR of Zebrafish ", fontsize=26)
plt.xlabel('Proportion of Weighted Occurance', fontsize=22)
plt.ylabel('Count', fontsize=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.vlines(df.loc['Mean']['Proportion of Weighted Occurance'], 0, 3,linestyle='dashed', label="mean")
plt.vlines(median, 0, 3,linestyle='dashed', label="median", color='blue')
plt.legend(fontsize=20)

```
